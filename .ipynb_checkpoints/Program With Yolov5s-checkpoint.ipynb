```python
import torch
import numpy as np
import cv2
from math import sqrt

def centre(x1, y1, x2, y2):
    Cx = x1 + (x2 - x1) // 2
    Cy = y1 + (y2 - y1) // 2
    return (Cx, Cy)

def distance(x1, y1, x2, y2):
    distance = sqrt((x1 - x2)**2 + (y1 - y2)**2)
    return distance

# Charger le modèle YOLOv5 pré-entraîné
modele = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)  # Utiliser le CPU

distanceAutorisee = int(input('Donnez la distance autorisée: '))
video = cv2.VideoCapture('videos/video1.mp4') # Vous pouvez changer la vidéo à utiliser

while video.isOpened():
    verif, cadre = video.read()
    if not verif:
        break
    resultats = modele(cadre)
    etiquettes, coordonnees = resultats.xyxyn[0][:, -1], resultats.xyxyn[0][:, :-1]
    largeur_cadre, hauteur_cadre = cadre.shape[1], cadre.shape[0]
    for i in range(len(coordonnees)):
        pers1 = coordonnees[i]
        if modele.names[int(etiquettes[i])] != modele.names[0]:
            continue
        if pers1[4] >= 0.2:
            pers1x1, pers1y1 = int(pers1[0] * largeur_cadre), int(pers1[1] * hauteur_cadre)
            pers1x2, pers1y2 = int(pers1[2] * largeur_cadre), int(pers1[3] * hauteur_cadre)
            distanceSuffisante = True
            for j in range(len(coordonnees)):
                if i == j:
                    continue # Pour éviter de vérifier la distance avec soi-même
                pers2 = coordonnees[j]
                if modele.names[int(etiquettes[j])] != modele.names[0]:
                    continue
                if pers2[4] >= 0.2:
                    pers2x1, pers2y1 = int(pers2[0] * largeur_cadre), int(pers2[1] * hauteur_cadre)
                    pers2x2, pers2y2 = int(pers2[2] * largeur_cadre), int(pers2[3] * hauteur_cadre)
                    (pers1cx, pers1cy) = centre(pers1x1, pers1y1, pers1x2, pers1y2)
                    (pers2cx, pers2cy) = centre(pers2x1, pers2y1, pers2x2, pers2y2)
                    if distance(pers1cx, pers1cy, pers2cx, pers2cy) < distanceAutorisee:
                        distanceSuffisante = False
                        break
            if not distanceSuffisante:
                cv2.rectangle(cadre, (pers1x1, pers1y1), (pers1x2, pers1y2), (0, 0, 255), 2)
                cv2.putText(cadre, modele.names[int(etiquettes[i])], (pers1x1, pers1y1 - 10), cv2.FONT_HERSHEY_PLAIN, 0.9, (0, 0, 255), 2)
            else:
                cv2.rectangle(cadre, (pers1x1, pers1y1), (pers1x2, pers1y2), (0, 255, 0), 2)
                cv2.putText(cadre, modele.names[int(etiquettes[i])], (pers1x1, pers1y1 - 10), cv2.FONT_HERSHEY_PLAIN, 0.9, (0, 255, 0), 2)   
    cv2.imshow("Vidéo", cadre)
    touche = cv2.waitKey(1)
    if touche == 27:
        break
video.release()
cv2.destroyAllWindows()
```

```python
import torch
import numpy as np
import cv2
from math import sqrt

def center(x1, y1, x2, y2):
    Cx = x1 + (x2 - x1) // 2
    Cy = y1 + (y2 - y1) // 2
    return (Cx,Cy)
def distance(x1, y1, x2, y2):
    distance = sqrt((x1 - x2)**2 + (y1 - y2)**2)
    return distance
# model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained = True).cuda() # cuda(): to use GPU instead of CPU
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)  # Removed .cuda() to use CPU

distanceAutorise = int(input('Donnerz la distance autorise: '))
video = cv2.VideoCapture('videos/video1.mp4') # you can change the video to use
while video.isOpened():
    check, frame = video.read()
    if not check:
        break
    results = model(frame)
    labels, cords = results.xyxyn[0][:, -1], results.xyxyn[0][:, :-1]
    x_shape, y_shape = frame.shape[1], frame.shape[0]
    for i in range(len(cords)):
        pers1 = cords[i]
        if model.names[int(labels[i])] != model.names[0]:
            continue
        if pers1[4] >= 0.2:
            pers1x1, pers1y1 = int(pers1[0] * x_shape), int(pers1[1] * y_shape) # cordinates of top left point of person 1
            pers1x2, pers1y2 = int(pers1[2] * x_shape), int(pers1[3] * y_shape) # cordinates of bottom right point of person 1
            enoughDistance = True
            for j in range(len(cords)):
                if i == j:
                    continue # To avoiding checking the distance between an object and itself
                pers2 = cords[j]
                if model.names[int(labels[j])] != model.names[0]:
                    continue
                if pers2[4] >= 0.2:
                    pers2x1, pers2y1 = int(pers2[0] * x_shape), int(pers2[1] * y_shape) # cordinates of top left point of person 2
                    pers2x2, pers2y2 = int(pers2[2] * x_shape), int(pers2[3] * y_shape) # cordinates of bottom right point of person 2
                    (pers1cx, pers1cy) = center(pers1x1, pers1y1, pers1x2, pers1y2) # cordinates of center of person 1
                    (pers2cx, pers2cy) = center(pers2x1, pers2y1, pers2x2, pers2y2) # cordinates of center of person 2
                    if distance(pers1cx, pers1cy, pers2cx, pers2cy) < distanceAutorise:
                        enoughDistance = False
                        break
            if not enoughDistance:
                cv2.rectangle(frame, (pers1x1, pers1y1), (pers1x2, pers1y2), (0, 0, 255), 2)
                cv2.putText(frame, model.names[int(labels[i])], (pers1x1, pers1y1- 10), cv2.FONT_HERSHEY_PLAIN, 0.9, (0, 0, 255), 2)
            else:            
                cv2.rectangle(frame, (pers1x1, pers1y1), (pers1x2, pers1y2), (0, 255, 0), 2)
                cv2.putText(frame, model.names[int(labels[i])], (pers1x1, pers1y1- 10), cv2.FONT_HERSHEY_PLAIN, 0.9, (0, 255, 0), 2)   
    cv2.imshow("Video", frame)
    key=cv2.waitKey(1)
    if key ==27:
        break
video.release()
cv2.destroyAllWindows()
```
